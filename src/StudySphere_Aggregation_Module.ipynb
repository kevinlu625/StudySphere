```python
import pandas as pd
import json
from collections import defaultdict
```

Class Selection

```python
# Define available classes
classes = ['Math', 'Physics', 'Chemistry', 'Biology', 'Computer Science']

# Display classes and get user input
selected_class = input(f"Select a class from {classes}: ")
while selected_class not in classes:
    selected_class = input("Invalid selection. Please choose again: ")

print(f"Class '{selected_class}' selected.")
```

Choose Exam Type

```python
# Define exam types
exam_types = ['Midterm', 'Final']

# Display options and get user input
selected_exam = input(f"Select exam type ({exam_types}): ")
while selected_exam not in exam_types:
    selected_exam = input("Invalid selection. Please choose again: ")

print(f"Exam type '{selected_exam}' selected.")
```

Post a Question

```python
# Function to post a question
def post_question(class_name, exam_type):
    question_text = input("Enter your question: ")
    return {
        'class': class_name,
        'exam_type': exam_type,
        'question': question_text,
        'answers': [],  # To hold suggested answers
        'votes': 0      # To track votes
    }

# Create a list to hold posted questions
questions = []

# Add a question to the list
question = post_question(selected_class, selected_exam)
questions.append(question)

print(f"Question posted: {question['question']}")
```

Suggest an Answer

```python
# Function to suggest an answer to a question
def suggest_answer(question):
    answer_text = input("Enter your answer suggestion: ")
    question['answers'].append({
        'text': answer_text,
        'votes': 0
    })

# Add a suggested answer to the first question
suggest_answer(questions[0])
print("Suggested answer added.")
```

Vote on Questions and Answers

```python
# Upvote a question
def upvote_question(question):
    question['votes'] += 1

# Upvote an answer
def upvote_answer(answer):
    answer['votes'] += 1

# Usage
upvote_question(questions[0])
print(f"Question votes: {questions[0]['votes']}")

upvote_answer(questions[0]['answers'][0])
print(f"Answer votes: {questions[0]['answers'][0]['votes']}")
```

Display Data

```python
# Display the list of questions with answers and votes
pd.DataFrame(questions)
```

```python
## Example displaying our sample data and using simulated JSON data

import pandas as pd
import json

# Load initial data from example_cards.csv
example_data = pd.read_csv('/Users/kerrykz/StudySphere/data/example_cards.csv')
print("Example Cards Data:")
display(example_data)

# Simulated JSON data for questions and votes
user_questions = [
    {
        "Username": "User1",
        "Question Type": "Multiple Choice",
        "Question": "What elements are in the periodic table?",
        "QuestionId": 1,
        "Answer": "Carbon, Mitochondria, Nucleus"
    },
    {
        "Username": "User2",
        "Question Type": "Short Answer",
        "Question": "What is the mitochondria?",
        "QuestionId": 2,
        "Answer": "The powerhouse of the cell."
    },
    {
        "Username": "User3",
        "Question Type": "Short Answer",
        "Question": "What algorithm can be used to sort a list?",
        "QuestionId": 3,
        "Answer": "Quick Sort"
    },
    {
        "Username": "User4",
        "Question Type": "Short Answer",
        "Question": "What algorithm can be used to sort a list?",
        "QuestionId": 3,
        "Answer": "Merge Sort"
    }
]

question_votes = [
    {"Username": "User5", "QuestionId": 3, "Upvote": True},
    {"Username": "User6", "QuestionId": 2, "Upvote": False}
]

answer_votes = [
    {"Username": "User7", "QuestionId": 3, "Answer": "Quick Sort"},
    {"Username": "User8", "QuestionId": 3, "Answer": "Merge Sort"}
]

# Aggregate Questions and Answers
questions_dict = {}
for item in user_questions:
    qid = item["QuestionId"]
    if qid not in questions_dict:
        questions_dict[qid] = {
            "Question": item["Question"],
            "Type": item["Question Type"],
            "Answers": [],
            "TotalVotes": 0,
            "UserVotes": {}
        }
    questions_dict[qid]["Answers"].append({"Text": item["Answer"], "Votes": 0, "Username": item["Username"]})

# Process Question Votes
for vote in question_votes:
    qid = vote["QuestionId"]
    if qid in questions_dict:
        if vote["Upvote"]:
            questions_dict[qid]["TotalVotes"] += 1
        else:
            questions_dict[qid]["TotalVotes"] -= 1

# Process Answer Votes
for vote in answer_votes:
    qid = vote["QuestionId"]
    answer_text = vote["Answer"]
    for answer in questions_dict[qid]["Answers"]:
        if answer["Text"] == answer_text:
            answer["Votes"] += 1

print("Aggregated Questions and Answers:")
for qid, data in questions_dict.items():
    print(f"Question ID {qid} - {data['Question']} (Type: {data['Type']})")
    print("Answers:")
    for answer in sorted(data["Answers"], key=lambda x: x["Votes"], reverse=True):
        print(f"  - {answer['Text']} (Votes: {answer['Votes']})")
```
